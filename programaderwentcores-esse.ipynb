```python

# SEgundo passo: gerar nodes.csv e edges.csv

import random

cluster_dic = {'AU': 1, 'BR': 2, 'CA': 3, 'CN': 4, 'DE': 5, 'EP': 6, 'FR': 7, 'GB': 8, 'IN': 9, 'JP': 10, 'KR': 11, 'LU': 12, 'RU': 13, 'SG': 14, 'TW': 15, 'US': 16, 'VN': 17, 'WO': 18, 'NL':19, 'TR':20, 'PL':21, 'ES':22, 'HK':23}

# extraindo depositantes de uma lista de documentos salvos em entrada.txt
filename = 'nodes.csv'
#if not os.path.exists(filename): 
saida = open(filename, 'w')
saida.write("id	label	description	x	y	cluster	weight<Links>	weight<Total link strength>	weight<Citations>	score<Journal impact factor>\n")
filename = 'edges.csv'
saida2 = open(filename, 'w')
#saida2.write("Source,Target,Type,Weight\n")
resposta_final = {}
id_resposta_final = {}
cluster_resposta_final = {}
id = 1
w, h = 20000, 20000
matrix = [[0 for x in range(w)] for y in range(h)] 

with open('entrada_new.txt','r') as file: #  geração do arquivo edges.txt
    for linha in file:
        partes = linha.split('*')
        numero = partes[0].strip()
        print(numero)
        country = numero[0:2]
        linha_depositantes = partes[1]

        depositantes = linha_depositantes.split(';')
        resultado=[]
        for elemento in depositantes:
            elemento = elemento.strip()
            resultado.append(elemento)
            print(elemento)
            if elemento in resposta_final:
                resposta_final[elemento]+=1
            else:
                resposta_final[elemento]=1 # contador com número de vezes que esse depositante aparece nos resultados gerais
                id_resposta_final[elemento]=id # isso será usado para criar o arquivo nodes.csv dos depositantes e respectivos ids
                cluster_resposta_final[elemento]=cluster_dic[country] # country da patente e não do depositante !
                id=id+1

            r = resposta_final[elemento]
            s = id_resposta_final[elemento]
            print(f'carga=({elemento})({r})(id={s})')

        #print(resposta_final)            
        temp = []
        for elemento in depositantes:
            elemento = elemento.strip()
            print(f'depositante={elemento}')
            temp.append(id_resposta_final[elemento]) # obtem o índice deste elemento em resposta_final
                    
        z = len(temp)
        print(f'temp={temp} z={z}')
        if z>1:
            for i in range(0,z):
                if (i+1)<=z:
                    for j in range(i+1,z):
                        print(f'i={i} j={j}')
                        imaior = temp[i]
                        imenor = temp[j]
                        print(f'imaior={imaior} imenor={imenor}')
                        if imaior>imenor:
                            matrix[imenor][imaior]+=1
                        else:
                            matrix[imaior][imenor]+=1
        
    z = len(resposta_final)
    for i in range(1,z+1):
        for j in range(i,z+1):
            k = matrix[i][j]
            if k>0: # somente mostre as conexoes que tiveram ao menos uma patente em comum
                saida2.write(f'{i}	{j}	{k}\n')
                    
print(resposta_final) 
print(id_resposta_final) 
print(cluster_resposta_final)

for elemento in id_resposta_final: # geração do arquivo nodes.txt
    id = id_resposta_final[elemento]
    weight = resposta_final[elemento]
    cluster = cluster_resposta_final[elemento]
    elemento = elemento.replace("[BR]","")
    z = len(elemento)
    if z<30:
        label = elemento
    else:
        label = elemento[0:8]+'..'+elemento[z-20:z]
    label = label.lower()
    # cluster = 1
    coordx = round(random.uniform(-1, 1), 4)
    coordy = round(random.uniform(-1, 1), 4)
    weight_total = weight
    weight_citations = weight
    score = 1
    saida.write(f'{id}	{label}	<table><tr><td>{elemento}</td></tr><tr><td>Peso: {weight}</td></tr></table>	{coordx}	{coordy}	{cluster}	{weight}	{weight_total}	{weight_citations}	{score}\n')
    
saida.close()
saida2.close()
print('Fim de processamento')
```
